```python
import numpy as np
import mlrose_hiive as mlr
import mlrose_hiive.runners as runners
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
# Exploration defining a new fitness function for rhc with tracking

eval_store = 0

def queens_peaks_with_eval(state):
    global eval_store
    eval_store += 1
    return mlr.Queens().evaluate(state)

problem = mlr.DiscreteOpt(length=100, fitness_fn=mlr.CustomFitness(queens_peaks_with_eval), max_val=2)
```

### Comparing All Functions

```python
eval_store = {
    'rhc': 0,
    'sa': 0,
    'ga': 0,
    'mimic': 0
}

eval_temp_store = 0

def queens_peaks_with_eval(state):
    global eval_temp_store
    eval_temp_store += 1
    return mlr.Queens().evaluate(state)

problem = mlr.DiscreteOpt(length=100, fitness_fn=mlr.CustomFitness(queens_peaks_with_eval), max_val=2, maximize=True)
```

```python
output_folder = 'Queens/compare_all'

# Fitness vs Iterations
iteration_range = np.arange(0,551, 50)

rhc = runners.rhc_runner.RHCRunner(problem=problem,
                    experiment_name='fitness_vs_iterations_rhc',
                    output_directory= output_folder,
                    seed=42,
                    iteration_list= iteration_range,
                    max_attempts=10,
                    restart_list=[0])

sa = runners.sa_runner.SARunner(problem=problem,
                    experiment_name='fitness_vs_iterations_sa',
                    output_directory= output_folder,
                    seed=42,
                    iteration_list= iteration_range,
                    max_attempts=10,
                    temperature_list=[1])

ga = runners.GARunner(problem=problem,
                  experiment_name='fitness_vs_iterations_ga',
                  output_directory=output_folder,
                  seed=42,
                  iteration_list=iteration_range,
                  max_attempts=10,
                  population_sizes=[150],
                  mutation_rates=[.5])

mimic = runners.MIMICRunner(problem=problem,
                  experiment_name='fitness_vs_iterations_mimic',
                  output_directory=output_folder,
                  seed=42,
                  iteration_list=iteration_range,
                  max_attempts=10,
                  population_sizes=[150],
                  keep_percent_list=[0.25],
                  use_fast_mimic=True)

runner_store = [['rhc',rhc], ['sa', sa], ['ga', ga], ['mimic', mimic]]

for i in runner_store:
    i[1].run()
    eval_store[i[0]] = eval_temp_store
    eval_temp_store = 0

print(eval_store['rhc'])
print(eval_store['sa'])
print(eval_store['ga'])
print(eval_store['mimic'])
    
```

```python
rhc_df = pd.read_csv(output_folder + '/' + 'fitness_vs_iterations_rhc/rhc__fitness_vs_iterations_rhc__run_stats_df.csv' )
sa_df = pd.read_csv(output_folder + '/' + 'fitness_vs_iterations_sa/sa__fitness_vs_iterations_sa__run_stats_df.csv' )
ga_df = pd.read_csv(output_folder + '/' + 'fitness_vs_iterations_ga/ga__fitness_vs_iterations_ga__run_stats_df.csv' )
mimic_df = pd.read_csv(output_folder + '/' + 'fitness_vs_iterations_mimic/mimic__fitness_vs_iterations_mimic__run_stats_df.csv' )
```

```python
plt.figure()
plt.plot(rhc_df['Iteration'], rhc_df['Fitness'], label='RHC')
plt.plot(sa_df['Iteration'], sa_df['Fitness'], label='SA')
plt.plot(ga_df['Iteration'], ga_df['Fitness'], label='GA')
plt.plot(mimic_df['Iteration'], mimic_df['Fitness'], label='MIMIC')

plt.title('Comparing All Search Algorithms: Fitness vs Iterations')
plt.xlabel('Iteration')
plt.ylabel("Fitness")
plt.legend()
plt.grid()
plt.show()
```
