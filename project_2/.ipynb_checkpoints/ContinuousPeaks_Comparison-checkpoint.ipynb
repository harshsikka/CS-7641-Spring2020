```python
import numpy as np
import mlrose_hiive as mlr
import mlrose_hiive.runners as runners
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
# Exploration defining a new fitness function for rhc with tracking

eval_store = 0

def continuous_peaks_with_eval(state):
    global eval_store
    eval_store += 1
    return mlr.ContinuousPeaks().evaluate(state)

problem = mlr.DiscreteOpt(length=100, fitness_fn=mlr.CustomFitness(continuous_peaks_with_eval), max_val=2)
```

```python
# Running RHC for OneMax
best_state, best_fitness, fitness_per_iteration = mlr.random_hill_climb(problem, max_iters=1000, random_state=42, curve=True)
```

```python
print('fitness_per_iteration: ' + str(fitness_per_iteration.shape))
```

### RHC curves for OneMax

```python
output_folder_rhc = 'onemax_rhc'

# Fitness vs Iterations
iteration_range = np.arange(0,551, 50)
rhc_fit_vs_it = runners.rhc_runner.RHCRunner(problem=problem,
                    experiment_name='fitness_vs_iterations',
                    output_directory= output_folder_rhc,
                    seed=42,
                    iteration_list= iteration_range,
                    max_attempts=10,
                    restart_list=[0])   
```

```python
rhc_fit_vs_it.run()

rhc_fit_vs_it_df = pd.read_csv(output_folder_rhc + '/' + 'fitness_vs_iterations/rhc__fitness_vs_iterations__run_stats_df.csv' )
plt.figure()
plt.plot(rhc_fit_vs_it_df['Iteration'], rhc_fit_vs_it_df['Fitness'])
plt.title('RHC on OneMax Problem: Iterations vs Fitness')
plt.xlabel('Iterations')
plt.ylabel("Fitness")
plt.legend()
plt.grid()
plt.show()
```

```python
# Fitness vs Restarts
iteration_range = np.arange(100)
rhc_fit_vs_re = runners.rhc_runner.RHCRunner(problem=problem,
                    experiment_name='fitness_vs_restarts',
                    output_directory= output_folder_rhc,
                    seed=42,
                    iteration_list= [120],
                    max_attempts=10,
                    restart_list=[5,10,15,20,25])   
```

### SA curves for OneMax

```python
output_folder_sa = 'onemax_sa'

# fitness vs iterations
iteration_range = np.arange(0,151, 10)
sa_fit_vs_it = runners.SARunner(problem=problem,
                  experiment_name='fitness_vs_iterations',
                  output_directory=output_folder_sa,
                  seed=42,
                  iteration_list=iteration_range,
                  max_attempts=10,
                  temperature_list=[1])
```

```python
sa_fit_vs_it.run()

sa_fit_vs_it_df = pd.read_csv(output_folder_sa + '/' + 'fitness_vs_iterations/sa__fitness_vs_iterations__run_stats_df.csv' )
plt.figure()
plt.plot(sa_fit_vs_it_df['Iteration'], sa_fit_vs_it_df['Fitness'])
plt.title('SA on OneMax Problem: Iterations vs Fitness')
plt.xlabel('Iterations')
plt.ylabel("Fitness")
plt.legend()
plt.grid()
plt.show()
```

```python
# Decays 

temp_store_geo = []
temp_store_arith = []
temp_store_exp = []


for i in range(1,201, 10):
    temp_store_geo.append(mlr.GeomDecay(init_temp=1.0, decay=0.99, min_temp=0.001).evaluate(i))

for i in range(1,201, 10):
    temp_store_arith.append(mlr.ArithDecay(init_temp=1.0, decay=0.0001, min_temp=0.001).evaluate(i))

for i in range(1,201, 10):
    temp_store_exp.append(mlr.ExpDecay(init_temp=1.0, exp_const=0.005, min_temp=0.001).evaluate(i))

# fitness vs geometric decay
sa_fit_vs_geo = runners.SARunner(problem=problem,
                  experiment_name='fitness_vs_geo',
                  output_directory=output_folder_sa,
                  seed=42,
                  iteration_list=[130],
                  max_attempts=10,
                  temperature_list=temp_store_geo)

# fitness vs arithmetic decay
sa_fit_vs_arith = runners.SARunner(problem=problem,
                  experiment_name='fitness_vs_arith',
                  output_directory=output_folder_sa,
                  seed=42,
                  iteration_list=[130],
                  max_attempts=10,
                  temperature_list=temp_store_arith)

# fitness vs exponential decay
sa_fit_vs_exp = runners.SARunner(problem=problem,
                  experiment_name='fitness_vs_exp',
                  output_directory=output_folder_sa,
                  seed=42,
                  iteration_list=[130],
                  max_attempts=10,
                  temperature_list=temp_store_exp)

sa_fit_vs_geo.run()
sa_fit_vs_geo_df = pd.read_csv(output_folder_sa + '/' + 'fitness_vs_geo/sa__fitness_vs_geo__run_stats_df.csv' )

sa_fit_vs_arith.run()
sa_fit_vs_arith_df = pd.read_csv(output_folder_sa + '/' + 'fitness_vs_arith/sa__fitness_vs_arith__run_stats_df.csv' )

sa_fit_vs_exp.run()
sa_fit_vs_exp_df = pd.read_csv(output_folder_sa + '/' + 'fitness_vs_exp/sa__fitness_vs_exp__run_stats_df.csv' )

plt.figure()
plt.plot(sa_fit_vs_geo_df['schedule_current_value'], sa_fit_vs_geo_df['Fitness'], label='Geometric')
plt.title('SA on OneMax Problem: Temp vs Fitness')
plt.xlabel('Temp')
plt.ylabel("Fitness")
plt.grid()
plt.show()

plt.figure()
plt.plot(sa_fit_vs_arith_df['schedule_current_value'], sa_fit_vs_geo_df['Fitness'], label='Arithmetic')
plt.title('SA on OneMax Problem: Temp vs Fitness')
plt.xlabel('Temp')
plt.ylabel("Fitness")
plt.grid()
plt.show()

plt.figure()
plt.plot(sa_fit_vs_exp_df['schedule_current_value'], sa_fit_vs_geo_df['Fitness'], label='Exponential')
plt.title('SA on OneMax Problem: Temp vs Fitness')
plt.xlabel('Temp')
plt.ylabel("Fitness")
plt.grid()
plt.show()

```

```python
# custom temperature modulation

# fitness vs exponential decay
sa_fit_vs_cus = runners.SARunner(problem=problem,
                  experiment_name='fitness_vs_temp',
                  output_directory=output_folder_sa,
                  seed=42,
                  iteration_list=[130],
                  max_attempts=10,
                  temperature_list=[.1,.2,.3,.4,.5, 1, 5, 10, 50, 100, 250])

sa_fit_vs_cus.run()
sa_fit_vs_cus_df = pd.read_csv(output_folder_sa + '/' + 'fitness_vs_temp/sa__fitness_vs_temp__run_stats_df.csv' )

plt.figure()
plt.semilogx(sa_fit_vs_cus_df['schedule_current_value'], sa_fit_vs_cus_df['Fitness'])
plt.title('SA on OneMax Problem: Temp vs Fitness')
plt.xlabel('Temp')
plt.ylabel("Fitness")
plt.grid()
plt.show()
```
